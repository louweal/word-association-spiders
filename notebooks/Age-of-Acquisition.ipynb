```python
import pandas as pd
import matplotlib.pylab as plt 
import seaborn as sns 

df = pd.read_csv('../data/processed/wordgame_20170831_ft.csv')
df['asd'] = df['forumID'].apply(lambda x:int((x%9)==0))

#df['AoA'] = df['AoA'].astype(float)
df = df[df['AoA'] != 10.1999]
len(df)
```

```python
sns.kdeplot(df[df['len2']<7].AoA, alpha=0.5,label="words with length <7")  
sns.kdeplot(df[df['len2']>11].AoA, alpha=0.5,label="words with length >11")  

plt.title('Distribution of a Age-of-Acquisition')
plt.xlabel('Age-of-Acquisition')
plt.ylabel('Percentage')
plt.legend(loc='upper right')
plt.show()
```

```python
print(df.tf.median())
```

```python
sns.kdeplot(df[df['tf']<0.013].AoA, alpha=0.5,label="words with below median frequency")  
sns.kdeplot(df[df['tf']>=0.013].AoA, alpha=0.5,label="words with above median frequency")  

plt.title('Distribution of a Age-of-Acquisition')
plt.xlabel('Age-of-Acquisition')
plt.ylabel('Percentage')
plt.legend(loc='upper right')
plt.show()

# Often used words have a really low AoA: ~4
```

```python
sns.kdeplot(df[df['user']=="cmackenzie4"].AoA, alpha=0.5,label="cmackenzie4 (TF)")  
sns.kdeplot(df[df['user']=="grimsby pete"].AoA, alpha=0.5,label="grimsby pete (TF)")  
sns.kdeplot(df[df['user']=="jj2"].AoA, alpha=0.5,label="jj2 (ECF)")  

plt.title('Distribution of a Age-of-Acquisition')
plt.xlabel('Age-of-Acquisition')
plt.ylabel('Percentage')
plt.legend(loc='upper right')
plt.show()

# jj2 is smartest? 
```
